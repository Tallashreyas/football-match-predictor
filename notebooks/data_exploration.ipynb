```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import glob

pd.set_option("display.max_columns", None)
```

```python
files = sorted(glob.glob("../data/raw/*.csv"))

df = pd.concat([pd.read_csv(file) for file in files], ignore_index=True)

df.head()
```

```python

df.head()
```

```python
print(df.columns.tolist())
```

```python
missing = df.isnull().sum()

missing = missing[missing > 0]

missing.sort_values(ascending=False)
```

```python
df["FTR"].value_counts()
df["FTR"].value_counts().plot(kind="bar")

plt.title("Match Results")
plt.xlabel("Result")
plt.ylabel("Count")

plt.show()
```

```python
teams = sorted(set(df["HomeTeam"]).union(df["AwayTeam"]))

print("Number of teams:", len(teams))

teams
```

```python
print("Average Home Goals:", df["FTHG"].mean())

print("Average Away Goals:", df["FTAG"].mean())
```

```python
df["TotalGoals"] = df["FTHG"] + df["FTAG"]

df.loc[df["TotalGoals"].idxmax()]
```

```python
df["Date"] = pd.to_datetime(df["Date"], dayfirst=True)

df = df.sort_values("Date")

df.head()
```

```python
columns = [
    "Date",
    "HomeTeam",
    "AwayTeam",
    "FTHG",
    "FTAG",
    "FTR",
    "HS",
    "AS",
    "HST",
    "AST",
    "HC",
    "AC",
    "HY",
    "AY",
    "HR",
    "AR",
]

clean_df = df[columns]

clean_df.head()
```

```python
clean_df.to_csv("../data/processed/clean_matches.csv", index=False)

print("Dataset saved!")
```
